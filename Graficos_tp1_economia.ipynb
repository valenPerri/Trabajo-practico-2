```python
import matplotlib.pyplot as plt

# Datos
cantidad_consumida = [0, 1, 2, 3, 4, 5]
utilidad_total = [0, 100, 180, 240, 280, 300]
utilidad_marginal = [0, 100, 80, 60, 40, 20]

# Graficar los datos
plt.figure(figsize=(10, 5))

plt.plot(cantidad_consumida, utilidad_total, label='Utilidad Total', marker='o')
plt.plot(cantidad_consumida, utilidad_marginal, label='Utilidad Marginal', marker='o')

plt.xlabel('Cantidad Consumida')
plt.ylabel('Utilidad')
plt.title('Utilidad Total y Utilidad Marginal en funci√≥n de la Cantidad Consumida')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
cantidad_consumida = [0, 1, 2, 3, 4, 5]
utilidad_total = [0, 100, 180, 240, 280, 300]

# Graficar los datos
plt.figure(figsize=(8, 5))

plt.plot(cantidad_consumida, utilidad_total, marker='o', color='blue')

plt.xlabel('Cantidad Consumida')
plt.ylabel('Utilidad Total')
plt.title('Utilidad Total en funci√≥n de la Cantidad Consumida')
plt.grid(True)

plt.show()
```
